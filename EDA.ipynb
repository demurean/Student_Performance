```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Make plots look nice
plt.style.use("seaborn-v0_8")
```

```python
# Path to your dataset
csv_path = "./data/Students_Performance_Dataset.csv"

try:
    df = pd.read_csv(csv_path)
    print("✅ Dataset loaded successfully!")
except FileNotFoundError:
    print(f"❌ Error: {csv_path} not found.")
    raise
```

```python
def create_categorical_chart(column_name):
    value_counts = df[column_name].value_counts()
    num_unique_values = len(value_counts)

    plt.figure(figsize=(3, 3))

    if num_unique_values <= 5:
        plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', radius = 1.0)
        plt.title(f'Pie Chart of {column_name}')
    else:
        value_counts.plot(kind='bar')
        plt.title(f'Bar Chart of {column_name}')
        plt.xlabel(column_name)
        plt.ylabel('Count')
        plt.xticks(rotation=45, ha='right')

    plt.tight_layout()
    plt.show()
```

```python
categorical_columns = [
    'Gender',
    'Department',
    'Extracurricular_Activities',
    'Internet_Access_at_Home',
    'Parent_Education_Level',
    'Family_Income_Level'
]

for col in categorical_columns:
    create_categorical_chart(col)
```
